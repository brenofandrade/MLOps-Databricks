```python
from databricks.sdk import WorkspaceClient
```

```python
w = WorkspaceClient()

sample_input = [
    {
        "sepal_length_cm": 5.1,
        "sepal_width_cm": 3.5,
        "petal_length_cm": 1.4,
        "petal_width_cm": 0.2
    }
]

prediction = w.serving_endpoints.query(
    name="predict_iris",
    dataframe_records=sample_input
)

display(prediction)
```

```python
print(
    'Input:',
    sample_input,
    '\nOutput: ',
    prediction.predictions
)
```
